```python
import os
import pandas as pd
```

```python
raw = pd.read_excel(
  '../../data/reference/qualifications_2021census.xlsx',
  sheet_name='quals_con'
).set_index([
            'ONSConstID', 
            'ConstituencyName', 
            'RegNationID', 
            'RegNationName' ]).drop(
    columns={'NatComparator', 
            'RN_pc', 
            'Nat_pc', 
            'ranking_total',
            'rank'}).rename(
    columns={'Con_num': 'value',
            'Con_pc': 'percent'})

raw.columns = raw.columns.str.strip()
raw
```

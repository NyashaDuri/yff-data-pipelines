```python
import pandas as pd
```

```python
data = pd.read_csv('../../data/raw/lfs_by_pcon.csv', parse_dates=['DATE'])
```

```python
filtered = data.loc[
  data.MEASURES_NAME == 'Variable',
  ['GEOGRAPHY_CODE', 'GEOGRAPHY_NAME', 'DATE', 'VARIABLE_NAME', 'MEASURES', 'MEASURES_NAME', 'OBS_VALUE', 'OBS_STATUS_NAME']
]
```

```python
def get_latest_available(data):
  # Calculate dates
  failsafe_date = data.DATE.min() - pd.Timedelta("1 days")
  latest_date = data.DATE.max()
  # Subset the input data
  test_data = data.loc[:, ['GEOGRAPHY_CODE', 'DATE', 'VARIABLE_NAME', 'OBS_VALUE']]
  # Construct a failsafe
  failsafe_data = test_data.loc[test_data.DATE == latest_date]
  failsafe_data.loc[:, ['DATE', 'OBS_VALUE']] = (failsafe_date, -1)
  test_data = pd.concat([ test_data, failsafe_data ])
  print(test_data.dropna(
          subset='OBS_VALUE'
      ))
  indexes = test_data.dropna(
          subset='OBS_VALUE'
      ).sort_values(
          'DATE', ascending=True
      ).groupby(
          ['GEOGRAPHY_CODE', 'VARIABLE_NAME']
      )['DATE'].idxmax()
  return data.loc[
      test_data.dropna(
          subset='OBS_VALUE'
      ).sort_values(
          'DATE', ascending=True
      ).groupby(
          ['GEOGRAPHY_CODE', 'VARIABLE_NAME']
      )['DATE'].idxmax()
  ]
  
latest_available = get_latest_available(data=filtered)
```
